```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8])
dummies1 = pd.get_dummies(df['Event_ID'], prefix='Event_ID')
dummies2 = pd.get_dummies(df['DOLocationID'], prefix='DOLocationID')

df = pd.concat([dummies1,dummies2,df["time"],df["TimeOfDay_Afternoon"],df["TimeOfDay_Evening"],df["TimeOfDay_Morning"],df["TimeOfDay_Night"],df["busyness_score"],df["busyness_score_event"]], axis=1)
df
```

```python
df.reset_index(drop=True, inplace=True)
X = df.drop('busyness_score_event', axis=1)
y = df['busyness_score_event']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```
